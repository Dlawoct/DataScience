```python
import torch
```

```python
#학습 데이터 생성
#y = 2x + 3x - 4
x_train = torch.FloatTensor([[1, 2], [3, 2], [3, 7], [1, 1], [1, 0]])
y_train = torch.FloatTensor([[4],[8],[23], [1], [-2]])

#w,b 초기화 및 Learning rate 설정
w = torch.randn(2, 1)
b = torch.randn(1, 1)
print(w)
print(b)
lr = 0.01

#반복 횟수 설정 및 requires_grad를 True로 설정
for epoch in range(3001):
  w.requires_grad_(True)
  b.requires_grad_(True)

  h = x_train @ w + b
  cost = ((h - y_train) ** 2).mean() # 평균

  #기울기 계산 및 w,b 업데이트
  cost.backward()
  with torch.no_grad(): #이미 계산한 기울기를 가지고 업데이트 하는 것이라 no_grad
    w = w - lr * w.grad
    b = b - lr * b.grad #b.grad -> cost를 b로 미분했을 때

    if epoch % 100 == 0:
      print('epoch: {}, cost: {:.6f}, w: {}, b: {}'.format(epoch, cost.item(), w.squeeze(), b))
```
