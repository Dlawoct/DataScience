```python
# 시그모이드 함수 정의
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 예측 함수 정의
def predict(X, w, b):
    return sigmoid(np.dot(X, w) + b)

# 손실 함수의 그래디언트 계산 함수
def compute_gradients(X, y, w, b):
    m = X.shape[0]
    y_pred = predict(X, w, b)  # 예측 값 계산
    dw = (1 / m) * np.dot(X.T, (y_pred - y))
    db = (1 / m) * np.sum(y_pred - y)
    return dw, db

# 가중치 및 절편 업데이트 함수
def update_weights(X, y, w, b, learning_rate):
    dw, db = compute_gradients(X, y, w, b)
    w -= learning_rate * dw
    b -= learning_rate * db
    return w, b

# 예제 데이터
X = np.array([[1, 1], [1, 2], [2, 3], [3, 3]])
y = np.array([0, 0, 1, 1])  # 이진 값

# 가중치와 절편 초기화
w = np.zeros(X.shape[1])
b = 0
learning_rate = 0.1

w, b = update_weights(X, y, w, b, learning_rate)

# 결과 출력
print(f'Updated weights (w): {w}')
print(f'Updated intercept (b): {b}')
```

```python
# 거리 구하기
def euclidean_distance(x, y):
  #if x.shape != y.shape:
        #raise ValueError("x와 y는 같은 shape을 가져야 합니다.")
    return np.sqrt(np.sum((x - y) ** 2))

def manhattan_distance(x, y):
    return np.sum(np.abs(x - y))

def cosine_distance(x, y):
    return 1 - np.dot(x, y) / (np.linalg.norm(x) * np.linalg.norm(y))

def hamming_distance(x, y):
    return np.sum(x != y)

def jaccard_distance(x, y) :
    intersection = np.sum(np.logical_and(x, y))
    union = np.sum(np.logical_or(x, y))
    return 1 - intersection / union

x = np.array([2, 3, 5, 1, 4])
y = np.array([-1, 3, 9,-3, 6])

# 거리 계산
print("Euclidean Distance:", euclidean_distance(x, y))
print("Manhattan Distance:", manhattan_distance(x, y))
print("Cosine Distance:", cosine_distance(x, y))
print("Hamming Distance:", hamming_distance(x, y))
print("Jaccard Distance:", jaccard_distance(x, y))
```

```python
# 지니 불순도
def gini_impurity(classes):
    total = sum(classes.values())
    impurity = 1.0
    for count in classes.values():
        p = count / total
        impurity -= p ** 2
    return round(impurity, 2) # 소수점

classes = {'A': 64, 'B': 95}
gini = gini_impurity(classes)
print("Gini Impurity:", gini)
# weighted avg는 더해서 / N
```

```python
# Numeric data
def gini_impurity(classes):
    total = sum(classes.values())
    impurity = 1.0
    for count in classes.values():
        p = count / total
        impurity -= p ** 2
    return impurity

def find_best_split(data, labels):
    # 데이터 정렬
    sorted_indices = np.argsort(data)
    sorted_data = data[sorted_indices]
    sorted_labels = labels[sorted_indices]

    unique_values = np.unique(sorted_data)
    best_gini = float('inf')
    best_split = None

    # 각 사이 값에 대해 Gini impurity 계산
    for i in range(len(unique_values) - 1):
        split_value = (unique_values[i] + unique_values[i + 1]) / 2

        # 분할
        left_indices = sorted_data <= split_value
        right_indices = sorted_data > split_value

        left_labels = sorted_labels[left_indices]
        right_labels = sorted_labels[right_indices]

        # 각 서브셋의 클래스 분포 계산
        left_classes = {label: np.sum(left_labels == label) for label in np.unique(labels)}
        right_classes = {label: np.sum(right_labels == label) for label in np.unique(labels)}

        # Gini impurity 계산
        left_impurity = gini_impurity(left_classes)
        right_impurity = gini_impurity(right_classes)

        # 전체 Gini impurity 계산 (가중 평균)
        total_impurity = (len(left_labels) * left_impurity + len(right_labels) * right_impurity) / len(labels)

        # 최적 분할 점 찾기
        if total_impurity < best_gini:
            best_gini = total_impurity
            best_split = split_value

    return best_split, best_gini

# 예시 데이터
data = np.array([6.3, 5.3, 7.2, 6.3, 9.8, 7.2, 6.3, 5.7, 8.0, 6.2])
labels = np.array(['x', 'o', 'x', 'o', 'o', 'x','o','o','x','x'])

best_split, best_gini = find_best_split(data, labels)
print(f"Best Split: {best_split}, Best Gini Impurity: {best_gini}")
```

```python
# k-means 중심점 찾기
def calculate_new_centroids(data_points, centroids):
    # 각 데이터 포인트에 대해 가장 가까운 중심점 찾기
    distances = np.linalg.norm(data_points[:, np.newaxis] - centroids, axis=2)
    closest_centroids = np.argmin(distances, axis=1)

    # 새로운 중심점 계산
    new_centroids = np.array([data_points[closest_centroids == k].mean(axis=0) for k in range(centroids.shape[0])])

    return new_centroids

# 예제 데이터 포인트와 초기 중심점
data_points = np.array([[1, 1], [1, 2], [1, 4], [1, 5], [2, 1], [4,4]])
initial_centroids = np.array([[2, 3], [3, 3]])

# 새로운 중심점 계산
new_centroids = calculate_new_centroids(data_points, initial_centroids)
print("새로운 중심점:", new_centroids)
```

```python
# k-means 갯수 구하기
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

# Iris 데이터셋 로드
iris = load_iris()
X = iris.data  # 특성 데이터 (꽃의 길이와 너비)

# SSE(총 제곱 오차)를 저장할 리스트
sse = []
k_values = range(1, 11)  # k 값을 1부터 10까지

# 각 k 값에 대해 K-means 수행
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)  # inertia_는 SSE를 반환

# Elbow Method를 시각화
plt.figure(figsize=(10, 6))
plt.plot(k_values, sse, marker='o')
plt.title('Elbow Method For Optimal k (Iris Dataset)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Cost (SSE - Sum of Squared Errors)')
plt.xticks(k_values)
plt.grid()
plt.show()

#data = pd.read_csv('your_dataset.csv')  # 실제 데이터셋의 파일 경로
#X = data[['feature1', 'feature2', 'feature3', ...]].values  # 필요한 특성을 선택
```
